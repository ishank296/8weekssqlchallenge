What is the customer count and percentage of customers who have churned rounded to 1 decimal place?

-  When customers cancel their Foodie-Fi service - they will have a churn plan record with a null price but their plan will continue until the end of the billing period.


How many customers have churned straight after their initial free trial - what percentage is this rounded to the nearest whole number?


```python
%%sql 
with not_trial_churned_customers as 
(
select distinct customer_id
from
subscriptions where 
plan_id not in (select plan_id from plans where plan_name in ('churn','trial'))
order by customer_id  
)
select s.customer_id,s.plan_id ,s.start_date,row_number() 
over(partition by s.customer_id order by s.start_date) as rnk
from subscriptions s
where customer_id not in (select customer_id from not_trial_churned_customers)
```

What is the customer count and percentage breakdown of all 5 plan_name values at 2020-12-31?


```python
%%sql
with customer_latest_plan as
(
select s.customer_id,
       s.start_date,
       p.plan_name,
       row_number() over(partition by customer_id order by start_date desc) as rnk 
from subscriptions s inner join plans p on p.plan_id = s.plan_id  
where start_date <='2020-12-31'
) ,
total_customer_cnt as (
select count(distinct customer_id) total_customers from subscriptions
where start_date <= '2020-12-31'
)
select plan_name, count(1),total_customers
from customer_latest_plan cross join total_customer_cnt
where rnk=1
group by plan_name,total_customers
```

How many customers have upgraded to an annual plan in 2020?


```python
%%sql
select count(1) from subscriptions
where plan_id in (select plan_id from plans where plan_name like '%annual%')
and extract(YEAR from start_date) = 2020
```

How many days on average does it take for a customer to an annual plan from the day they join Foodie-Fi?


```python
%%sql
with customer_join_date as (
select customer_id,
       min(start_date) join_date
from subscriptions
group by customer_id
),
customers_annual_plan as (
select customer_id,
       start_date
from subscriptions where 
   plan_id in (select plan_id from plans where plan_name like '%annual%')
)
select avg(a.start_date-b.join_date) average_conv_time 
from customers_annual_plan a
inner join customer_join_date b 
on a.customer_id = b.customer_id
```
